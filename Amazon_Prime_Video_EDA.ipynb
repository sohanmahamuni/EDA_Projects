# **Univariate Analysis**

#### Chart - 1

````python
# Chart - 1 visualization code
# Genre Distribution
plt.figure(figsize=(12, 6))
genre_counts = titles['genres'].explode().value_counts().head(10)
sns.barplot(x=genre_counts.index, y=genre_counts.values, palette="viridis")
plt.xticks(rotation=45)
plt.title('Top 10 Most Common Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
````

##### 1. Why did you pick the specific chart?

A bar chart is best suited for categorical data like genres, as it clearly displays the frequency of each genre.

##### 2. What is/are the insight(s) found from the chart?

* The most common genres include Drama, Comedy, and Documentation.Other genres like Horror, Thriller, and Romance appear less frequently in comparison.

* The presence of empty brackets ([]) indicates that some content lacks a genre classification, which could be a data quality issue.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes, understanding which genres are most in demand helps in content acquisition and production.

* If Drama and Comedy are overrepresented, other genres like Horror, Thriller, and Romance may be underrepresented, leading to missed opportunities in diverse audience segments.
* If genre diversity is too low, it may result in viewer fatigue, pushing them to competing platforms with a broader variety of content.

#### Chart - 9

````python
# Chart - 9 visualization code
# IMDb Rating Across Top 15 genress (Box Plot)

# Ensure 'genres' is processed correctly
titles['genres'] = titles['genres'].apply(lambda x: x.split(', ') if isinstance(x, str) else [])

# Explode the genres column
genre_ratings = titles.explode('genres')

# Filter top genres based on the number of movies (adjust limit as needed)
top_genres = genre_ratings['genres'].value_counts().nlargest(15).index
filtered_df = genre_ratings[genre_ratings['genres'].isin(top_genres)]

# Sorting genres by median IMDb rating for better visualization
sorted_genres = filtered_df.groupby("genres")["imdb_score"].median().sort_values().index
filtered_df["genres"] = pd.Categorical(filtered_df["genres"], categories=sorted_genres, ordered=True)

# Plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='genres', y='imdb_score', data=filtered_df, palette="coolwarm")

# Improving readability
plt.xticks(rotation=45, ha="right")  # Rotate labels slightly
plt.xlabel('Genre')
plt.ylabel('IMDb Rating')
plt.title('IMDb Ratings Across Top Genres (Filtered & Sorted)')

plt.show()
````

##### 1. Why did you pick the specific chart?

A box plot is ideal for visualizing the distribution of IMDb ratings across different genres. It effectively shows:

##### 2. What is/are the insight(s) found from the chart?

* Drama and European genres have higher median IMDb ratings compared to others.
* Comedy has a strong rating distribution, following Drama, exhibits a high median rating, with a well-balanced distribution.
* Thriller and Action genres have a wider spread of ratings, meaning they include both high and low-rated movies.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

* Thriller and Action genres show inconsistency: Some movies succeed, while others fail dramatically. Selective production and better storytelling are necessary to ensure high-quality releases.
* Many low-rated outliers in these genres suggest that some movies are not meeting audience expectations, potentially leading to negative reviews and poor business performance.
